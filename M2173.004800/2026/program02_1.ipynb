## First bit on sequences: lists
We will discuss lists in more detail later, but the following will help general understanding. We define a list using the square brackets `[` (open) and `]` (close), and put in the elements in the list separated by commas:


```python
lst1 = [1, 2, 3]
lst2 = lst1 
```

The book explains that we should view variable assignment as a kind of ''sticky'' note that we attach. In the above example, we create a list `[1,2,3]` and we put two "sticky" notes on them. Let's see how this works. We remark here that lists are mutable: we can change elements as follows:

```python
lst1[0] = 5
lst1
```

We see that we have modified the first entry, element 0. However, if we now look at the other list, we see

```python
lst2
```

So this has changed as well, since both `lst1` and `lst2` were references to the same object. To be precise, this is not the case with every type of variable.

```python
x = 5
y = x
print(x, y)
```

If we just change `x`, we see that the values change as follows.

```python
x = 6
print(x, y)
```

So in this case, `x` and `y` are not just references. As a rule of thumb, variables related to "big" data types, such as lists, tend to be references, whereas variables related to basic values, such integers and floats.

Sometimes we just want to modify a copy, and copy the original for other purposes. In that case, we can do the following.

```python
import copy #this module is always installed.

lst1 = [1, 2, 3]
lst2 = copy.deepcopy(lst1)

print(lst1)
print(lst2)
```

So we have identical copies. There are other ways, too, which can have advantages (this method is slow). We just mention this method as it works for all types.

For completeness, if we change one copy, the other doesn't change:

```python
lst1[0] = 5
print(lst1)
print(lst2)
```

## Graphics and plotting
In order to show more interesting examples, we import a plotting module. Instead of trying to cover all the options we have, we will just give a simple example, where we plot a regular grid of points.

To do this, we need a little bit more on lists. Lists can be extended using the `append` operation. We will create a list consisting of a grid of 5 by 3 points.

```python
x_coors = [] #empty list
y_coors = [] #empty list
for i in range(5):
    for j in range(3):
        x_coors.append(i)
        y_coors.append(j)
        print(i, j)

print(x_coors)
print(y_coors)
```

We make a scatter plot using the imported library. This has to be once before using it.
`conda install matplotlib` in the correct environment.

```python
import matplotlib.pyplot as plt #this module is not installed by default.
```

```python
plt.scatter(x_coors, y_coors)
plt.show()
```

This grid has $5 \cdot 3=15$ points. We can hence use the numbers $0,1,\ldots, 14$ to index each point:
If $i$ denotes the horizontal coordinate and $j$ the vertical coordinate, then we can index as
$$
index = 3 \cdot i +j.
$$
We can also invert this process, for example using the above lists:

```python
def get_coor(idx):
    return (x_coors[idx], y_coors[idx])

get_coor(5)
```

We can also use division with remainder to do this.

```python
def get_coor2(idx):
    return (idx //3, idx %3 )

for idx in range(15):
    print(get_coor2(idx) == get_coor(idx) )
```

### Logic operations
The simplest type is the `bool`, or boolean. It can only attain the values `False` and `True`. As such, it can be encoded by a single switch, or bit. It plays a role in logic and of course in the flow control of a program. 

The following operations exist:
`not`, `or`, `and`, `^`
The last one is called _exclusive or_ or _xor_.

The logic table is as follows:
`not False=True`, `not True=False`

`False or False = False`, `False or True = True`, `True or True=True`

`False and False = False`, `False and True = False`, `True and True=True`

`False ^ False = False`, `False ^ True = True`, `True ^ True=False`

We have also operations on other types of variables. These are
`==` and `!=` 
These check equality or inequality of objects (for numerical types there is also `>`, `<`, etc)

In addition, there is `is`. It is probably simplest to explain these using some examples.


```python
print(5==5)
print(4==5)

print(4!=5)
print("ab" == "ab")
```

To explain `is`, let's consider three lists.

```python
lst1 = [1,2]
lst2 = [1,2]
lst3 = lst1
# element-wise, these lists are the same. However, as objects
print(lst1 == lst2)
print(lst1 == lst3) 

print(lst1 is lst2) #different objects
print(lst1 is lst3) #same object
```
