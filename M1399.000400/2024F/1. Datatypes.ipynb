0. **Numeric types**: integer, float, boolean
    
0. **Sequence types**:  string, tuple, range, list

(These are all objects.)

Like Java and R, Python doesn't require to explicitly declare variable types.

Assignment is done using: `(variable name) = (value)`.

Variable names can be composed of a~z, A~Z, 0~9 and `_`, but should not start with numbers!

3. immutable vs mutable sequences
>immutable sequence: string, tuple, range / mutable sequence: list

### Lists

```python
#Lists are similar to tuples, except that they are mutable !
#Also, lists cannot be used as elements of sets while tuples can.

l=[3,7,5,'a']

l[0], l[-1], l[1:2]
```

```python
print(len(l))

print(l+[8,'b'])

#Be careful !! [1,2,3]+[1,1,1] is not [2,3,4], but [1,2,3,1,1,1]
```

```python
'A' in l
```

```python
l[0]=4 # This works. 'list' object supports item assignment.
```

Lists are objects!!

```python
l.append(10); print(l)

del l[3]; print(l)

l.reverse(); print(l)

l2=[3,4,2,5]

l2.sort(); print(l2)

l2.pop(0); print(l2)

l2.remove(5);print(l2)
```

## 3. Sets

```python
a={2,2,3,4,5,'a'}
print(a)

a=[2,2,3,4,5,5]
a=set(a)
print(a)
a.add(6)
print(a)
a.remove(5)
print(a)
```

We can use subtractions for sets!

```python
numbers=set(range(10))
evens={0,2,4,6,8}
odds=numbers-evens
print(odds)
```

Sets also allow union(`|`) and intersections(`&`)!

```python
f={2,4,6,8}
g={4,8,12}
print(f|g)
print(f&g)
```
